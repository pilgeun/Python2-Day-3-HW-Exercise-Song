## Exercise 1: 
Create an abstract class called Vehicle with 
- color, make, and year private attributes 

and abstract method called 
- get_info() which returns all information
- honk() which returns a string of honking sound

Create Car, Boat, and Bicycle class which is a child class of the Vehicle class.  Then, create an instance of each child class and call get_info() method to print out its information in a user friendly format. Add other methods as you like.

```python
from abc import ABC, abstractmethod

class Vehicle :
    """Vehicle class with getters and setters for private attributes"""
    def __init__(self, color, make, year) :
        """Initialize color, make and year private attributes"""
        self.__color = color
        self.__make = make
        self.__year = year    
    
    def get_color(self) :
        return self.__color
    
    def get_make(self) :
        return self.__make
    
    def get_year(self):
        return self.__year
    
    @abstractmethod
    def get_info(self) :
        """will return vehicle information"""
        pass
    
    @abstractmethod
    def honk(self) :
        """will return vehicle sound"""
        pass


class Car(Vehicle):
    """Car information and sound"""
    def get_info(self) :
        return f"{self.get_color()} {self.get_make()} car produced in {self.get_year()}"
    
    def honk(self) :
        return f"{self.get_make()} makes a Honk"
    
class Boat(Vehicle) :
    """Boat information and sound"""
    def get_info(self) :
        return f"{self.get_color()} {self.get_make()} boat manufactured in {self.get_year()}"
    
    def honk(self) :
        return f"{self.get_make()} goes Awooga"
    
class Bicycle(Vehicle) :
    """Bicycle information and sound"""
    def get_info(self) :
        return f"{self.get_color()} {self.get_make()} bicycle made in {self.get_year()}"
    
    def honk(self) :
        return f"{self.get_make()} says Click"


car_ferrari = Car("Red", "Enzo Ferrari", 2002)
boat_johnson = Boat("White", "Johnson", 1980)
bicycle_bmx = Bicycle("Blue", "Proline", 2021)
    
print(car_ferrari.get_info())
print(car_ferrari.honk())
print()

print(boat_johnson.get_info())
print(boat_johnson.honk())
print()

print(bicycle_bmx.get_info())
print(bicycle_bmx.honk())
print()
```

## Exercise 2: 
Create an abstract class called Person with name and address attributes, and print_info() abstract method. 
1. Build an Employee class inherited from Person class that 
    - stores hired date and salary attributes 
    - print_info() method that will be used to print out all information about an employee
    - a method to calculate employee's monthly pay where federal tax is 15%, MD tax is 4.5%, and a local tax is 3%.
    
2. Build a Customer class also inherited from Person class that
    - stores balance attribute
    - print_info() method that will be used to print out all information about a customer


Build any method you need but apply encapsulation at all possible. Then, write a program to show your classes work with multiple employees and customers. 

CHALLENGE!!!
Create an Address class and use it in Person class

```python
from abc import ABC, abstractmethod

class Person :
    """"""
    def __init__(self, name, address) :
        """Initialize name and address attributes"""
        self.name = name
        self.address = address
    
    @abstractmethod
    def print_info(self) :
        """will return Person information"""
        pass


class Employee(Person) :
    """"""
    def __init__(self, name, address, hireDate, salary) :
        """Initialize name, address, hireDate and salary attributes"""
        super().__init__(name, address)
        self.hireDate = hireDate
        self.salary = salary
        
    def print_info(self) :
        """return name, address, hireDate and salary"""
        return f"Name: {self.name} \nAddress: {self.address} \nHire date: {self.hireDate} \nSalary: ${self.salary*0.775:.2f}/month\n"
    
    
class Customer(Person) :
    """"""
    def __init__(self, name, address, balance) :
        """Initialize name, address and balance attributes"""
        super().__init__(name, address)
        self.balance = balance
        
    def print_info(self) :
        """return name, address and balance"""
        return f"Name: {self.name} \nAddress: {self.address} \nBalance: ${self.balance}\n"
                
# these addresses are actually based in California 
emp_daniel = Employee("Daniel", "Escalon Dr, San Fierro, MD 91436", '11-04-1961', 8000)
emp_johnny = Employee("Johnny", "Burbank Blvd, Los Santos, MD 91335", '10-20-1962', 1000)
cus_miguel = Customer("Miguel", "Burbank Blvd, Los Santos, MD 91335", 100)
cus_robby = Customer("Robby", "Lankershim Blvd, Las Venturas, MD 91602", 85)
cus_sam = Customer("Sam", "Escalon Dr, San Fierro, MD 91436", 500)
                 
print(emp_daniel.print_info())                 
print(emp_johnny.print_info())
print(cus_miguel.print_info())
print(cus_robby.print_info())
print(cus_sam.print_info())
```
